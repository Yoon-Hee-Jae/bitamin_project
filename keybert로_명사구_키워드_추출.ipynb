```python
# 계산수가 너무 복잡
for i in range(len(df1)):
  # 뉴스요약

  tokenizer =  PreTrainedTokenizerFast.from_pretrained("ainize/kobart-news")
  model = BartForConditionalGeneration.from_pretrained("ainize/kobart-news")

  inputs_ids = tokenizer.encode(df1['content'][i], return_tensors="pt")

# kobart가 정해둔 데이터 크기의 한계가 있어서 텍스트가 긴 뉴스같은 경우는 잘라야할듯
# 모델 입력 최대값 구하는 코드
  max_input_length = model.config.max_position_embeddings

  inputs_ids = inputs_ids[:, :max_input_length]

  summary_text_ids = model.generate(
    input_ids = inputs_ids,
    bos_token_id = model.config.bos_token_id,
    eos_token_id = model.config.eos_token_id,
    length_penalty = 2.0,
    max_length = 128,
    min_length = 32,
    num_beams = 4
)
  # 요약한 값을 summary라는 새 열에 저장
  # summary_text_ids[0]가 맞음 i 아님
  df1['summary'][i] = tokenizer.decode(summary_text_ids[0], skip_special_tokens=True)

      # 키워드 추출
  nouns = okt.nouns(df1['content'][i])
  text = ' '.join(nouns)
  #keywords = kw_model.extract_keywords(text, keyphrase_ngram_range=(1, 1), stop_words=None, top_n=10)
  # 하나에서 두개까지의 모든 키워드 가능성 고려 n-gram 파라미터 조정으로 바꿀 수 있음
  keyword = kw_model.extract_keywords(text, keyphrase_ngram_range=(1,2), top_n=10)


    # 데이터프레임에 키워드 할당
  df1.at[i, 'keywords'] = keyword
```
