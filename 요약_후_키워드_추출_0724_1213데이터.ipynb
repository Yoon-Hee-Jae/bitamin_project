```python
from keybert import KeyBERT
from kiwipiepy import Kiwi
from transformers import BertModel
from transformers import PreTrainedTokenizerFast, BartForConditionalGeneration
import pandas as pd
```

```python
# 형태소 정리
# kiwi가 knolpy보다 개선된 모델은 맞음 더 빠르고 정확하다고 함
# knolpy에는 Mecab, Hannanum, Kkma, Okt(Open Korean Text), Komoran으로 다양하며 mecab의 경우 windows에서는 사용 불가

# kiwi 형태소 분석기 사용 (kobert요약 후 keybert)
df = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/news_0713-0714.xlsx")
```

```python
df['keywords'] = None
df['summary'] = None
df
```

```python
# 명사 추출 함수
def noun_extractor(text):
    results = []
    result = kiwi.analyze(text)
    for token, pos, _, _ in result[0][0]:
        if len(token) != 1 and pos.startswith('N') or pos.startswith('SL'):
            results.append(token)
    return results

# 형태소 분석기
kiwi = Kiwi()

# 키워드 추출 모델
model2 = BertModel.from_pretrained('skt/kobert-base-v1')
kw_model = KeyBERT(model2)
```

```python
# 계산수가 너무 복잡
for i in range(30):
  # 뉴스요약

  tokenizer =  PreTrainedTokenizerFast.from_pretrained("ainize/kobart-news")
  model = BartForConditionalGeneration.from_pretrained("ainize/kobart-news")

  inputs_ids = tokenizer.encode(df['content'][i], return_tensors="pt")

# kobert가 정해둔 데이터 크기의 한계가 있어서 텍스트가 긴 뉴스같은 경우는 잘라야할듯
# 모델 입력 최대값 구하는 코드
  max_input_length = model.config.max_position_embeddings

  inputs_ids = inputs_ids[:, :max_input_length]

  summary_text_ids = model.generate(
    input_ids = inputs_ids,
    bos_token_id = model.config.bos_token_id,
    eos_token_id = model.config.eos_token_id,
    length_penalty = 2.0,
    max_length = 128,
    min_length = 32,
    num_beams = 4
)
  # 요약한 값을 summary라는 새 열에 저장
  # summary_text_ids[0]가 맞음 i 아님
  df['summary'][i] = tokenizer.decode(summary_text_ids[0], skip_special_tokens=True)

      # 키워드 추출
  kiwi.analyze(df['summary'][i])
  nouns = noun_extractor(df['summary'][i])
  text = ' '.join(nouns)
  keywords = kw_model.extract_keywords(text, keyphrase_ngram_range=(1, 1), stop_words=None, top_n=10)

    # 데이터프레임에 키워드 할당
  df.at[i, 'keywords'] = keywords
```

```python
# ----------------------------------------------------------------------------------------------
# keybert 대신 tf-idf를 통해서 키워드 추출 - kiwi
```

```python
# 위에서 요약한 부분까지는 그대로 사용하고 keyword만 초기화시켜주기
df3 = df[['date','title','content','press','link','summary']] # 형태소 분석 o
df4 = df[['date','title','content','press','link','summary']] # 형태소 분석 x
df3['keywords'] = None
df4['keywords'] = None

df3[:10]
```

```python
# 형태소 분석기 지정
kiwi = Kiwi()
```

```python
# 키워드 추출 모델 "tf-idf"
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# 명사만 남겨둔 새로운 열 생성
df3['content_noun'] = None
for i in range(10):
  # 형태소 분석
  kiwi.analyze(df3['content'][i]) # 요약문의 명사를 고르는게 아니라 tf-idf는 문서 전체를 보고 키워드를 뽑기 때문에 원문 전체를 형태소 분석 해주어야 함
  nouns = noun_extractor(df3['content'][i])
  text = ' '.join(nouns)
  df3['content_noun'][i] = text

df3.head(10)
```

```python
# 본격 tf-idf를 활용한 키워드 분석
# 키워드 추출 모델 지정
tfidfv = TfidfVectorizer()

# 몇번째 행까지 키워드 분석을 진행할지 지정 후 벡터화 시킴
tfidf_matrix = tfidfv.fit_transform(df3['content_noun'][:10])

for i in range(10):
    doc_tfidf = tfidf_matrix[i].toarray().flatten() # 벡터화시킨 것을 다시 문자로
    sorted_indices = doc_tfidf.argsort()[::-1] # tf-idf값이 높을수록 중요한 키워드인데 높은 순으로 정렬
    top_keywords = [tfidfv.get_feature_names_out()[idx] for idx in sorted_indices[:5]]  # 상위 5개 단어만 추출
    # 키워드를 저장할 빈 리스트 생성
    keywords = []
    keywords.append(top_keywords)
    # 각 행에 알맞은 키워드 부여
    df3['keywords'][i] = keywords
print(df3['keywords'][:10])

# 상장이라는 단어를 잡은 것으로 보아 keybert를 통해서 키워드를 추출한 것보다 tf-idf를 통해서 키워드를 추출하는 것이 더 적절해보임
```

```python
# 형태소 분석을 하게 되면 문장의 단어들이 너무 자세하게 나누어져서 유상증자와 같은 단어를 잘 못잡을 가능성이 있음
# 따라서 형태소 분석을 진행하지 않고 키워드 추출만을 활용

tfidfv = TfidfVectorizer()

# 몇번째 행까지 키워드 분석을 진행할지 지정 후 벡터화 시킴
tfidf_matrix = tfidfv.fit_transform(df4['content'][:10])

for i in range(10):
    doc_tfidf = tfidf_matrix[i].toarray().flatten() # 벡터화시킨 것을 다시 문자로
    sorted_indices = doc_tfidf.argsort()[::-1] # tf-idf값이 높을수록 중요한 키워드인데 높은 순으로 정렬
    top_keywords = [tfidfv.get_feature_names_out()[idx] for idx in sorted_indices[:5]]  # 상위 5개 단어만 추출
    # 키워드를 저장할 빈 리스트 생성
    keywords = []
    keywords.append(top_keywords)
    # 각 행에 알맞은 키워드 부여
    df4['keywords'][i] = keywords
print(df4['keywords'][:10])


# 중요한 단어라고 불릴 수 있는 키워드는 형태소 분석을 시행하지 않는 것이 더 좋아보임
# 형태소 분석 x > tf-idf를 통해서 키워드 10개 추출 > 숫자만 있는 경우 삭제 > 형태소분석을 통해 키워드에 붙은 조사 처리 > 최조 키워드 추출
```
